```python
import cv2
import glob
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
import random
from tqdm.notebook import tqdm
np.random.seed(1)

path = glob.glob('s/leaf.jpg',recursive=True)
len(path)
orig = np.array([np.asanyarray(Image.open(img)) for img in path])
orig.shape
plt.figure(figsize=(9,9))
for i, img in enumerate(orig[0:16]):
    plt.subplot(4,4,i+1)
    plt.xticks([])
    plt.yticks([])
    plt.grid(False)
    plt.imshow(img)
plt.suptitle("Original", fontsize=20)
plt.show()
```
